```python
#loading in train data

default_na = 0

#list of providers
train = pd.read_csv('Train-1542865627584.csv')

#contains beneficiary KYC details like health conditions, region they belong to
train_bene = pd.read_csv('Train_Beneficiarydata-1542865627584.csv').fillna(default_na)

#provides insights about the claims filed for those patients who are admitted in the hospitals
train_inpatient = pd.read_csv('Train_Inpatientdata-1542865627584.csv').fillna(default_na)

#provides details about the claims filed for those patients who visit hospitals and not admitted in it
train_outpatient = pd.read_csv('Train_Outpatientdata-1542865627584.csv').fillna(default_na)
```

```python
train_benecln = train_bene[['IPAnnualReimbursementAmt','OPAnnualReimbursementAmt']][train_bene[['IPAnnualReimbursementAmt','OPAnnualReimbursementAmt']] >= 0].fillna(value = {'IPAnnualReimbursementAmt': train_bene['IPAnnualReimbursementAmt'].mean(),'OPAnnualReimbursementAmt': train_bene['OPAnnualReimbursementAmt'].mean()})
train_bene['IPAnnualReimbursementAmt'] = train_benecln['IPAnnualReimbursementAmt']
train_bene['OPAnnualReimbursementAmt'] = train_benecln['OPAnnualReimbursementAmt']
train_bene.describe()
```

```python
train_temp = pd.concat([train_inpatient,train_outpatient], ignore_index = True)
train_temp
```

```python
#Create Age column to the dataset --not working (yet) a lot of DOD is None so difficult
from datetime import timedelta

#(pd.to_datetime(trn_df['ClaimStartDt'],format='%Y-%m-%d') - pd.to_datetime(trn_df['DOB'],format='%Y-%m-%d'))/timedelta(days=365)

train_bene['DOB'] = pd.to_datetime(train_bene['DOB'] , format = '%Y-%m-%d')
train_bene['DOD'] = pd.to_datetime(train_bene['DOD'],format = '%Y-%m-%d',errors='ignore')
#train_bene['Age'] = round((train_bene['DOD'] - train_bene['DOB'])/timedelta(days=365))

test_bene['DOB'] = pd.to_datetime(test_bene['DOB'] , format = '%Y-%m-%d')
test_bene['DOD'] = pd.to_datetime(test_bene['DOD'],format = '%Y-%m-%d',errors='ignore')
# test_bene['Age'] = round((test_bene['DOD'] - test_bene['DOB'])/timedelta(days=365))
```
