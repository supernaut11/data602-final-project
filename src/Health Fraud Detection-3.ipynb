```python
import numpy as mp
import pandas as pd
import os
pd.set_option('display.max_columns', None)  
```

```python
# Loading in all of the training data

default_na = "None"

# List of providers
train_providers = pd.read_csv('Train-1542865627584.csv')

# Contains beneficiary KYC details like health conditions, region they belong to
train_bene = pd.read_csv('Train_Beneficiarydata-1542865627584.csv')

# Provides insights about the claims filed for those patients who are admitted in the hospitals
train_inpatient = pd.read_csv('Train_Inpatientdata-1542865627584.csv')

# Provides details about the claims filed for those patients who visit hospitals and not admitted in it
train_outpatient = pd.read_csv('Train_Outpatientdata-1542865627584.csv')
```

```python
# Add a flag to each of the data sets to indicate whether they came from the inpatient vs. outpatient frames
train_inpatient['In/Out'] = 'In'
train_outpatient['In/Out'] = 'Out'
```

```python
# Use the concat operation to create a union of the two data sets. This automatically handles any missing columns between the two sets, so no need
# to manually add missing columns to the outpatient data first.
train_df = pd.concat([train_inpatient,train_outpatient], ignore_index = True)
train_df
```

```python
# Do a left outer join in case there are any claims that don't have any corresponding beneficiaries defined in the beneficiary data set
train_df = train_df.merge(train_bene, how='left', on='BeneID')

# TODO: Are there any claims that don't have corresponding beneficiary definitions?

train_df
```

```python
# Code and data Cleanup below. 
```

```python
# Turns In and Out into a boolean value where 0=outpatient and 1=inpatient
train_df['In/Out'] = train_df['In/Out'].apply(lambda x: 1 if x== "In" else 0)

train_df['AttendingPhysician'] = train_df['AttendingPhysician'].apply(lambda ap: ap.split('PHY')[-1] if not pd.isnull(ap) else ap)
train_df['OperatingPhysician'] = train_df['OperatingPhysician'].apply(lambda op: op.split('PHY')[-1] if not pd.isnull(op) else op)
train_df['OtherPhysician'] = train_df['OtherPhysician'].apply(lambda op: op.split('PHY')[-1] if not pd.isnull(op) else op)
```

```python
# Build list of diagnosis code columns, because hard-coding makes me sad
diagnosis_code_cols = [f'ClmDiagnosisCode_{x+1}' for x in range(10)]

# Strip ASCII characters from all of the claim diagnosis codes.
import string
for col in diagnosis_code_cols:
    train_df[col] = train_df[col].apply(lambda c: str(c).strip(string.ascii_letters) if not pd.isnull(c) else c)

# Sanity check that all of the claim codes are now integers
assert sum([len([x for x in train_df[col] if not pd.isnull(x) and not str(x).isdigit()]) for col in diagnosis_code_cols]) == 0

train_df
```

```python
# Once we have all of our features in place, we should merge with the potential fraud data set. This will let us set up 
# independent/dependent variables for the decision tree to examine.
```

```python
# Use a left outer join so we don't throw away providers that might not have claims.
train_providers = train_providers.merge(train_df, how='left', on='Provider')
```

```python
train_providers
```

```python
# Now lets clean up the data some more and take some of those unnecessary leading characters out of the entries
```

```python
# Removes PRV prefix from Provider ID
train_providers['Provider'] = train_providers['Provider'].apply(lambda y: y.split('PRV')[1] if not pd.isnull(y) else y)
# Removes BENE prefix from BeneID --may not be necessary since I don't think this code will be fed into the tree
train_providers['BeneID'] = train_providers['BeneID'].apply(lambda z: z.split('BENE')[1] if not pd.isnull(z) else z)

train_providers
```

```python
#Create Age column to the dataset --not working (yet) a lot of DOD is None so difficult
from datetime import timedelta

#(pd.to_datetime(trn_df['ClaimStartDt'],format='%Y-%m-%d') - pd.to_datetime(trn_df['DOB'],format='%Y-%m-%d'))/timedelta(days=365)

train_providers['DOB'] = pd.to_datetime(train_providers['DOB'] , format = '%Y-%m-%d')
train_providers['DOD'] = pd.to_datetime(train_providers['DOD'],format = '%Y-%m-%d')
train_providers['Age'] = round((train_providers['DOD'] - train_providers['DOB'])/timedelta(days=365))
```

```python
# Calculates age of person at the time they make the claim
train_providers['Age_at_Claim'] = round((pd.to_datetime(train_providers['ClaimStartDt'],format='%Y-%m-%d') - pd.to_datetime(train_providers['DOB'],format='%Y-%m-%d'))/timedelta(days=365))
```
